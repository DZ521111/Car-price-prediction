```python
# import models from sklearn

from sklearn.model_selection import *
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_feature, Y_feature, test_size = 0.19)
```

```python
# using decision tree and random forest

from sklearn.ensemble import *
```

```python
rf = RandomForestRegressor()

# hyperparameters

n_estimators = [int(tree) for tree in np.linspace(start = 80, stop = 1500, num = 20)]
print(n_estimators)
```

```python
# other parameters for random forest algo are

max_features = ["auto", "sqrt"]

# max number of levels
max_depth = [int(level) for level in np.linspace(5, 60, num = 10)]

# min sample for split
min_sample_split = [2, 8, 11, 17, 120]

# min sample of leaf node
min_sample_leaf = [1, 3, 8, 13]
```

```python
# use RandomizedSearchCV to find out best parameters

random_para = {
                "n_estimators": n_estimators,
                "max_features": max_features,
                "max_depth": max_depth,
                "min_samples_split": min_sample_split,
                "min_samples_leaf": min_sample_leaf
              }

print(random_para)
```

```python
# choose best para among of above parameter
# cv = cross validation 

rf_para = RandomizedSearchCV(estimator = rf, param_distributions = random_para, scoring = "neg_mean_squared_error", n_iter = 15, cv = 5, verbose = 2, random_state = 42, n_jobs = 1)
```

```python
rf_para.fit(X_train, Y_train)
```

```python
prediction = rf_para.predict(X_test)
```

```python
prediction
```
